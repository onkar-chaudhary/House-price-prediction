# Seperating features and label for model building

```python
housing = train_set.drop('MEDV', axis = 1)
housing_label = train_set['MEDV'].copy()
```

# Missing values

```python
# Three ways to tackle missing values
```

```python
# 1 - Remove row of the missing value
housing.dropna(subset = ['RM'])
```

```python
# 2 - Remove the whole column containing missing values
housing.drop('RM', axis = 1)
```

```python
# 3 - Setting missing values to min, max or median of that column
median = housing['RM'].median()
housing['RM'].fillna(median)
```

```python
# We can do Option 3 using an imputer as well. This helps us to generalise this step for all columns
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'median')
imputer.fit(housing)
```

```python
imputer.statistics_
```

```python
X = imputer.transform(housing)
housing_tr = pd.DataFrame(X, columns = housing.columns)
```

```python
housing_tr.describe()
```

# Creating a pipeline & Feature scaling

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'median')),
    ('std_scaler', StandardScaler())
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
```

```python
housing_num_tr
```

# Model building

## 1. Linear regression

```python
from sklearn.linear_model import LinearRegression

model_lr = LinearRegression()
model_lr.fit(housing_num_tr, housing_label)
```

```python
# Checking the model on some data
some_data_lr = housing.iloc[:5]
some_labels_lr = housing_label.iloc[:5]

prepared_data_lr = my_pipeline.transform(some_data_lr)
```

```python
model_lr.predict(prepared_data_lr)
```

```python
list(some_labels_lr)
```

## Evaluating linear regression

```python
def print_scores(rmse_scores):
    print("Scores: ", rmse_scores)
    print("Mean: ", rmse_scores.mean())
    print("Standard deviation: ", rmse_scores.std())
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
housing_predictions_lr = model_lr.predict(housing_num_tr)
```

```python
# calculating RMSE
mse_lr = mean_squared_error(housing_label, housing_predictions_lr)
rmse_lr = np.sqrt(mse_lr)
```

```python
rmse_lr
```

```python
# calculating cross validation score
from sklearn.model_selection import cross_val_score
scores_lr = cross_val_score(model_lr, housing_num_tr, housing_label, scoring = 'neg_mean_squared_error', cv = 10)
rmse_scores_lr = np.sqrt(-scores_lr)
```

```python
print_scores(rmse_scores_lr)
```
